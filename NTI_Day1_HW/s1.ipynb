```python
students = [
    {"name": "Ahmed", "score": 90, "attendance_rate": 0.85},
    {"name": "Mona", "score": 40, "attendance_rate": 0.90},
    {"name": "Ali", "score": 70, "attendance_rate": 0.75}
]

for student in students:
    if student["score"] > 50 and student["attendance_rate"] > 0.8:
        print(f"{student['name']}: passed")
    else:
        print(f"{student['name']}: failed")
```
