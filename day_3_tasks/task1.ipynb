```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_acc = np.array([0.5, 0.65, 0.75, 0.82, 0.88, 0.91, 0.93, 0.95, 0.96, 0.97, 0.98, 0.985])
test_acc = np.array([0.48, 0.62, 0.72, 0.78, 0.83, 0.85, 0.87, 0.88, 0.89, 0.90, 0.90, 0.91])

train_error = np.array([0.5, 0.35, 0.25, 0.18, 0.12, 0.09, 0.07, 0.05, 0.04, 0.03, 0.02, 0.015])
test_error = np.array([0.52, 0.38, 0.28, 0.22, 0.17, 0.15, 0.13, 0.12, 0.11, 0.10, 0.10, 0.095])

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(train_acc, 'b-', label='Train Accuracy', linewidth=2)
plt.plot(test_acc, 'r-', label='Test Accuracy', linewidth=2)
plt.xlim(0, 11)
plt.ylim(0.4, 1.0)

plt.title('Accuracy vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(train_error, 'g-', label='Train Error', linewidth=2)
plt.plot(test_error, 'm-', label='Test Error', linewidth=2)

plt.xlim(0, 11)
plt.ylim(0, 0.6)

plt.title('Error vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
