```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("my_file.csv")
```

```python
df.head
```

```python
df.drop(columns=['Rank', 'Peak', 'All Time Peak', 'Ref.'], inplace=True)
```

```python
df.head(20)
```

```python
df.columns = df.columns.str.replace('\xa0', ' ').str.strip()
```

```python
df['Actual gross'] = df['Actual gross'].str.replace(r'[$,\[\]a-zA-Z]', '', regex=True).astype(float)
df['Adjusted gross (in 2022 dollars)'] = df['Adjusted gross (in 2022 dollars)'].str.replace(r'[$,\[\]a-zA-Z]', '', regex=True).astype(float)

print(df.head(20))
```

```python
df['Artist'] = df['Artist'].str.replace(r'[†‡*]|\[.*?\]', '', regex=True).str.strip()
df['Tour title'] = df['Tour title'].str.replace(r'[†‡*]|\[.*?\]', '', regex=True).str.strip()

df.head(20)
```

```python
df[['Start_Year', 'End_Year']] = df['Year(s)'].str.replace(r'\[.*?\]', '', regex=True).str.split(r'[–-]', expand=True)
```

```python
df.head(20)
```

```python
df['End_Year'] = df['End_Year'].fillna(df['Start_Year'])
```

```python
df.head(20)
```

```python
df = df.drop(columns=['Year(s)'])
```
